```python
list = ['banana','apple','orange',123,True];
#Access Items
print(list[1])

#Negative Indexing
# Negative indexing means start from the end
print(list[-2]);

#Range of Indexes
#You can specify a range of indexes by specifying where to start and where to end the range.
# print(list[1:3]) OR assign it with new variable. The original list will be unchanged.
newList = list[1:3];
print(newList)

#This example returns the items from the beginning to, but NOT including, "4th Index":
print(list[:4])

#This example returns the items from index2 to end:
print(list[2:])

#Range of Negative Indexes
#Specify negative indexes if you want to start the search from the end of the list:
print(list[-4:-2])

#Check if Item Exists
if "orange" in list:
    print("yes it is");
else:
    print("not it is")
    
# Python - Change List Items
# Change Item Value
# To change the value of a specific item, refer to the index number:
list[1] = 'Mango';
print(list)

#Change a Range of Item Values
#To change the value of items within a specific range, define a list with the new values, and refer to the range of index numbers where you want to insert the new values:
list[0:2] = ['tomatoes','potatoes']
print(list)

# Note: The length of the list will change when the number of items inserted does not match the number of items replaced.

#Python - Add List Items
#To add an item to the end of the list, use the append() method:

list.append("Sarmad")
print(list)

# Insert Items
# The insert() method inserts an item at the specified index:
# list.insert(3) = 'YES' Syntax Error
list.insert(3,'yes')
print(list)

#Extend List
# To append elements from another list to the current list, use the extend() method.
list2 = ["abc","xyz",123,False];
list3 = [];
list.extend(list2)
print("Extended List ")
print(list);
# how to insert all elements in newly empty list / array
print("List3 is ");
for i in list:
    list3.append(i)
print(list3)

#Python - Remove List Items

#remove()
#The remove() method removes the specified item.

list3.remove("Sarmad")
print("Sarmad is removed");
print(list3)

#Remove Specified Index
#pop()

list3.pop(3);
#yes removed
print(list3);
#If you do not specify the index, the pop() method removes the last item.

#The del keyword also removes the specified index:
del list3[0] #tomatoes gone
print(list3);

#The del keyword can also delete the list completely.
# del list3 # complete deleted
#print(list3) #NameError: name 'list3' is not defined
#but if we wanna no to delete completely but we wanna delete its all items, we use clear() method.
#It clears all items then that arry will become empty

# list3.clear();
# print(list3) # [] empty


# Python - Loop Lists

# Loop Through a List
print("Loop starts here")
for i in list3:
    print(i)
    
#Loop Through the Index Numbers
print("Printing using index numbers")
for i in range(len(list3)):
    # print(i) 0 1 2 ... 7
    print("Index ",i , "=",list3[i])
    
#Using a While Loop
print("Printing using while loop")
count=0
while count < len(list3):
    print(list3[count])
    # count++ not working
    # count+=1; # working
    count = count + 1; #working
    
# Python - Sort Lists    
# List objects have a sort() method that will sort the list alphanumerically, ascending, by default:
newlist = ["orange", "mango", "kiwi", "pineapple", "banana"]
print("Before sorting newlist")
print(newlist)
print("After sorting newlist")
newlist.sort()
print(newlist)

listNums = [10,4,2,54,32];
listNums.sort();
print(listNums);

#Sort Descending
listNums.sort(reverse = True)
print(listNums);

# Reverse Order
nums = [1,2,3,4,5]
nums.reverse();
print(nums);
```
