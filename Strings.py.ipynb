```python
#Multiline Strings

multiLine = '''
Hello, 

I am Sarfraz

Thanks'''

print(multiLine)

#Strings are Arrays
print("Strings are Arrays")
print(multiLine[1])

#Looping Through a String
print("Looping Through a String")
for i in multiLine:
    print(i)
#String Length
print("String Length");
print(len(multiLine))

#Check String
print("Check String");

print("Sarmad" in multiLine)

#Use it in an if statement:
if "Sarfraz" in multiLine:
    print("Sarfraz is there!")
else:
    print("Sarfraz is not there");
    
#Check if NOT

if "Sarmad" not in multiLine:
        print("Sarmad in not there!")
else:
    print("Sarmad is there!");
    
#Python - Slicing Strings

txt = "Hello World!"

print(txt[2:5])
print("Slice From the Start")
print(txt[:5])
print("Slice To the End")
print(txt[4:])
print("Negative Indexing")
# Use negative indexes to start the slice from the end of the string: 
print(txt[-7:-2])

#Python - Modify Strings
print("Python - Modify Strings");

#Upper Case
print("Upper Case")
a = 'sarfraz Rasheed';
upper = a.upper()
print(upper);

#Lower Case
print("Lower Case")
print(a.lower())

#Capital Case
print("First Letter Capital")
print(a.capitalize())

#Remove Whitespace
print("With Whitespace")
whiteSpace = '    Hello      '
print(whiteSpace);
print("Without Whitespace")
print(whiteSpace.strip());

#Replace String
print("Replace String");
print(a.replace("sarfraz","Tahir"))

#Split String
print("Split String");
#it converts string to array
converted = a.split()
print(type(converted));
print(converted);

print("Replacing sarfraz with Tahir")
for i in converted:    
#     if i == 'sarfraz': or
        if
        i = "Tahir"
    print(i)

```
